```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso,ElasticNet
from sklearn.metrics import mean_squared_error ,r2_score
```

```python
 X,y =make_regression(n_samples=500,n_features=10,noise=15,random_state=42)#synthetic regression data generate karta hai.
# n_samples → data points ka number (rows)

# n_features → features ka number (columns)

# noise → data me randomness add karta hai (realistic banane ke liye)

# random_state → same data repeatable generate karne ke liye
```

```python
#train testsplit
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size =0.1,random_state=42)
```

```python

#Ridge → coefficients small kar do, but zero nahi

#Lasso → coefficients small + kuch zero → feature selection

#ElasticNet → Ridge + Lasso ka combo → correlated features + selection#Ridge Regresion

## Ridge Regression
ridge_model =Ridge(alpha=1.0)
ridge_model.fit(X_train,y_train)
y_pred_ridge =ridge_model.predict(X_test)


#find the mean squared error and r2_score
mse_ridge =mean_squared_error(y_test,y_pred_ridge)
r2_ridge=r2_score(y_test,y_pred_ridge)
print(f"Ridge Regression -MSE:{mse_ridge},R2:{r2_ridge}")
```

```python
#Lasso Regesion
lasso_model =Lasso(alpha=0.1)
lasso_model.fit(X_train,y_train)
y_pred_lasso =lasso_model.predict(X_test)

#find themen squared error and r2 score

mse_lasso =mean_squared_error (y_test,y_pred_lasso)
r2_lasso =r2_score(y_test,y_pred_lasso)
print(f"lasso Regression MSE :{mse_lasso},R2: {r2_lasso}")
```

```python
#Elastic Regrseion
elastic_model =ElasticNet(alpha=0.5,l1_ratio=0.5)
elastic_model.fit(X_train,y_train)
y_pred_elastic=elastic_model.predict(X_test)

#find the mean squared error and r2 score


mse_elastic=mean_squared_error(y_test,y_pred_elastic)
r2_elastic=r2_score(y_test,y_pred_elastic)
print(f"Elastic Regression -MSE {mse_elastic},R2:{r2_elastic}")
```
