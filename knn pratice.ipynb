```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
#weight And price
x=np.array([
    [201,150],
    [206,999],
    [188,200],
    [197,599] 
])
#label categories(0-cheap,1-expensive)
y=np.array([0,1,0,1])

#creatcategories(0-cheap,1-expanisve)
model=KNeighborsClassifier(n_neighbors =3)#k-3
model.fit(x,y)

#predict one plus with weight 184g and price $300
predicition=model.predict([[1184,300]])
print(predicition)#Output[0](cheap)


#interpret the prediction
category = "expensive" if predicition[0] == 1 else "cheap"
print(f"The predicted category for the phone is: {category}")
```
