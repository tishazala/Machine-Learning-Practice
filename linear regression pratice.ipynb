```python
import numpy  as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#data 
X=np.array([8,10,19]).reshape(-1,1)
y=np.array([10,13,16])#pizza  price(depend variable)
```

```python
#create  and train  the model
#Y  = mX+b(linear equation)
model =LinearRegression()
model.fit(X,y)
```

```python
#get slop(m)
#get  the intercetp(b)
m= model.coef_[0]#slope
b=model.intercept_
print(f"slop:{m:.1f}")
print(f"Intercept:{b:.2f}")
```

```python
#predict price for 20 incge -price 
x_new =np.array([20]).reshape(-1,1)
y_predict =model.predict(x_new)
print(f"predicted price for 20 inch pizza ={y_predict[0]:.2f}")

```

```python
#plot the data  point
plt.scatter(X,y,color ='blue',label='data point')
plt.plot(X,model.predict(X),color ='red',label='Regression Line')
plt.xlabel("Diameter (inch)")
plt.ylabel("Price $")
plt.title("pizza Price Prediction")
plt.legend()
          
           
```

```python
#2 example
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
data ={
    'Size':[1000,1500,2000,2200,2800],
    'Bedroom':[2,3,4,5,6],
    'Price':[150,220,280,285,290]
}

```

```python
df=pd.DataFrame(data)
print(df)
X=df[['Size','Bedroom']]
y=df['Price']
```

```python
#Step 2 Fit the model
model=LinearRegression()
model.fit(X,y)
print("Intercept(B0)", model.intercept_)# b
print("Coefficient:(B1,B2)",model.coef_)# M =cofficient
```

```python

#predict the  values 1600 sqft and 3 bedroom
new_house=np.array([[1600,3]])
predict_price =model.predict(new_house)
print(f"Predict price for 1600 drt and 3 bedroom:{predict_price}")
```

```python
#step 4 evaluate the model
y_pred =model.predict(X)
mse =mean_squared_error(y,y_pred) #mse
print(f"Mean Squared Eroor :{mse:.2f}")
```

```python
r2=r2_score(y,y_pred)
print(f"R-squared :{r2:.2f}")#rmses
```
