```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
#Load dataset
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
df.head()
```

```python
X=df.drop("target",axis=1)
y=df["target"]
```

```python
##use =StandarScaler
scaler=StandardScaler()
X_scler=scaler.fit_transform(X)
```

```python
#apply PCA
pca=PCA(n_components=2)
X_pca =pca.fit_transform(X_scler)
print("Variance Ration=",pca.explained_variance_ratio_)
```

```python
#convet into Data FRame -data
pca_df=pd.DataFrame(data=X_pca,columns=['pcal','pca2'])
pca_df['target']=y
```

```python


# use = StandardScaler
scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
# apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaler)
print("Variance Ratio = ",pca.explained_variance_ratio_)


# convert into DataFrame -> data
pca_df = pd.DataFrame(data=X_pca,columns=['pca1','pca2'])
pca_df['target'] = y# pca plot
plt.figure(figsize=(8,6))
sns.scatterplot(
    data = pca_df,
    x = "pca1",
    y = "pca2",
    hue = "target",
    palette="viridis",
    s=10
)
plt.title("PCA")
plt.xlabel("PC 1")
plt.ylabel("pc 2")
plt.show()
```
